we can see salary column has some ouliers we need to fix this

```python
# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = df1['Salary'].quantile(0.25)
Q3 = df1['Salary'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter outliers
outliers = df1[(df1['Salary'] < lower_bound) | (df1['Salary'] > upper_bound)]
non_outliers = df1[(df1['Salary'] >= lower_bound) & (df1['Salary'] <= upper_bound)]
```

```python
outliers['Salary'].count()
```

we find out that the salary outliers is approx. 10% of our data so we keeping this to get more insight if possible

```python
#Find the gender diversity across organisation

gender_distribution = df1['Sex'].value_counts(normalize=True) * 100
gender_distribution = gender_distribution.round(2)
print(gender_distribution)
```

```python
gender_distribution.plot(kind='pie', autopct='%1.1f%%', colors=['lightblue', 'pink'], title='Overall Gender Diversity')
plt.ylabel('')  # Remove y-label for better aesthetics
plt.show()
```

```python
#Find the gender diversity department wise

dept_gender_distribution = df1.groupby('Department')['Sex'].value_counts(normalize=True).unstack() * 100
dept_gender_distribution = dept_gender_distribution.round(2)
print(dept_gender_distribution)
```

```python
# Plot donut charts for each department's gender diversity
departments = df1['Department'].unique()
colors = ['lightblue', 'pink']  # Customize colors

fig, axes = plt.subplots(1, len(departments), figsize=(10, 5))

for i, dept in enumerate(departments):
    dept_gender_distribution = df1[df1['Department'] == dept]['Sex'].value_counts()
    axes[i].pie(dept_gender_distribution, labels=dept_gender_distribution.index, autopct='%1.1f%%', colors=colors, startangle=90, wedgeprops={'width': 0.3})
    
    # Add a circle in the center to make it look like a donut chart
    centre_circle = plt.Circle((0, 0), 0.70, fc='white')
    fig.gca().add_artist(centre_circle)

    axes[i].set_title(dept)
    axes[i].set_aspect('equal')

plt.suptitle('Gender Diversity by Department')
plt.tight_layout()
plt.show()
```

From the above chart we can assume that most of the departmet as well as organization is more on less well gender diversified.

```python
df2 ['price'] = pd.to_numeric(df2['price'], errors='coerce')
sns.set_theme(rc={"figure.figsize": (20, 6)})
sns.histplot(df2['price'], kde=False, bins=30)
```

```python
print("Skewness: %f" % df2['price'].skew())
print("Kurtosis: %f" % df2['price'].kurt())
```

As the price column has skewness has value >1, it is posetivly skewed.

 Also high value of kurosis also indicate it has good amount of outlier

```python
#using IQR method to detect and remove outliers
q1 = df2['price'].quantile(0.25)
q3 = df2['price'].quantile(0.75)
iqr = q3 - q1

lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

New_df=df2[(df2['price'] >= lower_bound) & (df2['price'] <= upper_bound)]
New_df
```

```python
sns.histplot(New_df['price'], kde=False, bins=30)
```

```python
# Finding Which Neighbour Group has Most costliest place to live in

sns.violinplot(x='neighbourhood_group',y='price',data=New_df)
plt.title("Neighbourhood_group wise price distributions")
```

As we can see Manhattan is most costliest in Neighbourhood_group in the list. 

Brooklyn and Queens have almost same price distribution.

Staten Island does not vary much as its price is almost fix like in between (50-100)

Bronx price is the cheapest as its price is less than 50

```python
# Find which type of room is in demand

New_df.groupby('room_type')['minimum_nights'].mean().plot(kind='barh',color='g')
plt.title('Min Stays in different room types listed on Airbnb ')
plt.xlabel('Minimum Night Stays')
plt.ylabel('Room Types')
```

From the above plot its clear that people like to stay in Entire home/apt compared to private and shared room

```python
# find Which type property has majority in which neighbour_group

plt.rcParams['figure.figsize'] = (8, 5)
ax= sns.countplot(y='room_type',hue='neighbourhood_group',data=New_df,palette='bright')

total = len(New_df['room_type'])
for p in ax.patches:
        percentage = '{:.1f}%'.format(100 * p.get_width()/total)
        x = p.get_x() + p.get_width() + 0.02
        y = p.get_y() + p.get_height()/2
        ax.annotate(percentage, (x, y))

plt.title('Count of each room types in entire Airbnb')
plt.xlabel('Percentage acquired in neighbour_group')
plt.xticks(rotation=90)
plt.ylabel('Room Types')

plt.show()
```

From Above plot it can be assumed that :

 1.Manhatten has most property of Entire Home/Apt

 2.Brooklyn is slightly ahead in case of property of Private room than Manhattan

 3.Shared room only available in Manhattan
 
 4.Queens has few property of Private room and Entire room/Apt

```python
# Find the relation Between type of room and their availability in different Neighbourhood_group
f,ax = plt.subplots(figsize=(15,8))
ax=sns.boxplot(x='neighbourhood_group',y='availability_365',data=New_df,palette="bright",hue='neighbourhood_group', dodge=False)
plt.title("Neighbourhood Group vs. Room Availabilty")
plt.xlabel('Neighbourhood groups')
plt.ylabel('Availability(365)')
plt.show()
```

From this plot it is visible that:
